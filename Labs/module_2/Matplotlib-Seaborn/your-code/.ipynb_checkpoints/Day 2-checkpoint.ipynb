```python
import numpy as np
```

```python
x1=np.arange(0,10,1) # same as range, but creates an array
```

```python
x1
```

```python
x2=np.linspace(0,5,10) # generate 10 numbers from 0 to 5 including both numbers.
x2
```

```python
plt.plot(x1,x1, label='linear')
```

```python
plt.plot(x1,x2**2, label='quadratic')
```

```python
plt.plot(x1,x1, label='linear')
plt.plot(x1,x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Linear vs Quadratic')
plt.legend()

plt.grid()

plt.show()
```

```python
fig, axes=plt.subplots(1,2)
axes[0].plot(x1,x1, label='linear')
axes[1].plot(x1,x1, label='linear')

axes[1].plot(x1,x2**2, label='quadratic')

axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

axes[1].set_xlabel('x label')
axes[1].set_ylabel('y label')

axes[0].set_title('Linear one')
axes[1].set_title('Quadratic one')



plt.show()
```

```python
x=np.random.randn(10)
y=np.random.randn(10)

plt.scatter(x,y)
```

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 45, 15]
explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df['X']=np.arange(0,10,1)
df['Y']=df.X**2
```

```python
df
```

```python
sns.lineplot('X','Y',data=df)
```

```python
df.Y.plot()
```

```python
sns.scatterplot('X','Y',data=df)
```

```python
df.plot.scatter()
#cant be built directly from DataFrame, since 2 arguments are required
```

```python
sns.lmplot('X','Y',data=df, fit_reg=True)
```
